# Interpretable AI practice


Did you notice that D0 depends quadratically on DELTA? It appears it is the case.

One can write: D0(MACH_BLUE, MACH_RED, ALT_BLUE, ALT_RED, DELTA) = A0 + A1 * DELTA + A2 * DELTA**2, where A0, A1, and A2 depends on MACH_BLUE, MACH_RED, ALT_BLUE, and ALT_RED. 

Therefore, we can once again transform our regression problem. If we predict the coefficients A0, A1 and A2, then we can predict the PK thanks to D0 and the sigmoid shape.

```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit

```

```python
df = pd.read_csv("dataset_practical_session.csv")
y=df["PK"]
X=df.drop("PK", axis=1)
```

```python
def sigmoid(d, d0,k):
    return 1/(1+np.exp(k*(d-d0)))

df_d0=pd.DataFrame(columns=["MACH_BLUE", "MACH_RED", "ALT_BLUE", "ALT_RED", "DELTA", "D0", "K"])

new_rows=[]

for x1 in set(df["MACH_BLUE"]):
    for x2 in set(df["MACH_RED"]):
        for x3 in set(df["ALT_BLUE"]):
            for x4 in set(df["ALT_RED"]):
                for x5 in set(df["DELTA"]):
                    mask=(
                        (df["MACH_BLUE"]==x1)&(df["MACH_RED"]==x2)&
                        (df["ALT_BLUE"]==x3)&(df["ALT_RED"]==x4)&
                        (df["DELTA"]==x5)
                    )
                    X_f=X[mask]
                    y_f=y[mask]
                    d_values=X_f["DISTANCE"].values
                    y_values=y_f.values

                    popt, _ = curve_fit(sigmoid, d_values, y_values, [np.median(d_values), 0])

                    d0, k = popt
                    new_rows.append({"MACH_BLUE": x1, "MACH_RED": x2, "ALT_BLUE": x3, "ALT_RED": x4, "DELTA": x5, "D0":d0, "K":k})

df_d0=pd.DataFrame(new_rows)
```

```python
df_d0
```

### Question

- Instead of working on the database X[MACH_BLUE, MACH_RED, ALT_BLUE, ALT_RED, DELTA] —> y[D0], create the dataset X[MACH_BLUE, MACH_RED, ALT_BLUE, ALT_RED] —> y[A0, A1, A2]. What is the size of this dataset? Is it better to work on this dataset? 

```python
y=df_d0["D0"]
X=df_d0.drop(["D0","K"], axis=1)
```

```python
def cuadratic(d, A0, A1,A2):
    return A0+A1*d+A2*d**2

df_Cuadr=pd.DataFrame(columns=["MACH_BLUE", "MACH_RED", "ALT_BLUE", "ALT_RED", "A0", "A1", "A2"])

new_rows=[]

for x1 in set(df_d0["MACH_BLUE"]):
    for x2 in set(df_d0["MACH_RED"]):
        for x3 in set(df_d0["ALT_BLUE"]):
            for x4 in set(df_d0["ALT_RED"]):
                    mask=(
                        (df_d0["MACH_BLUE"]==x1)&(df_d0["MACH_RED"]==x2)&
                        (df_d0["ALT_BLUE"]==x3)&(df_d0["ALT_RED"]==x4)
                    )
                    X_f=X[mask]
                    y_f=y[mask]
                    d_values=X_f["DELTA"].values
                    y_values=y_f.values

                    popt, _ = curve_fit(cuadratic, d_values, y_values, [0, 0, 0])

                    A0, A1, A2 = popt
                    new_rows.append({"MACH_BLUE": x1, "MACH_RED": x2, "ALT_BLUE": x3, "ALT_RED": x4, "A0":A0, "A1":A1, "A2":A2})

df_Cuadr=pd.DataFrame(new_rows)
```

```python
df_Cuadr
```

```python
import matplotlib.pyplot as plt
import numpy as np

df_plot = df_d0[
    (df_d0["MACH_BLUE"] == 0.5) &
    (df_d0["MACH_RED"] == 1) &
    (df_d0["ALT_BLUE"] == 10) &
    (df_d0["ALT_RED"] == 0)
]

x_data = df_plot["DELTA"].values
y_data = df_plot["D0"].values

row_params = df_Cuadr[
    (df_Cuadr["MACH_BLUE"] == 0.5) &
    (df_Cuadr["MACH_RED"] == 0.5) &
    (df_Cuadr["ALT_BLUE"] == 0) &
    (df_Cuadr["ALT_RED"] == 0)
].iloc[0]

A0, A1, A2 = row_params["A0"], row_params["A1"], row_params["A2"]

x_fit = np.linspace(min(x_data), max(x_data), 200)
y_fit = A0 + A1*x_fit + A2*(x_fit**2)

plt.scatter(x_data, y_data, label="Real data", color="blue")
plt.plot(x_fit, y_fit, label="Adjust", color="red")
plt.xlabel("DELTA")
plt.ylabel("D0")
plt.legend()
plt.grid(True)
plt.show()
```

## Test regression



```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler

y = df_Cuadr[["A0", "A1", "A2"]]
x = df_Cuadr.drop(["A0", "A1", "A2"],axis=1)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

minmax_scaler = MinMaxScaler()
X_train_scaled = minmax_scaler.fit_transform(X_train)
X_test_scaled = minmax_scaler.transform(X_test)
```

### LinearRegression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score

regr = LinearRegression()

regr.fit(X_train_scaled, y_train)
print(regr.score(X_test_scaled, y_test))

```

```python
import time

start = time.time()
y_pred = regr.predict(X_test_scaled)
end = time.time()
inference_time = (end - start) / len(X_test_scaled)

print(f"time taken to do prediction {end - start}")

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2  = r2_score(y_test, y_pred)

print(f"R² (score): {r2:.4f}")
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
```

### Polynimal

```python
from sklearn.preprocessing import PolynomialFeatures

pr = PolynomialFeatures(degree=2)

X_train_poly = pr.fit_transform(X_train_scaled)
X_test_poly  = pr.transform(X_test_scaled)
lr_2 = LinearRegression()
lr_2.fit(X_train_poly, y_train)
```

```python
import time

start = time.time()
y_pred = lr_2.predict(X_test_poly)
end = time.time()
inference_time = (end - start) / len(X_test_scaled)

print(f"time taken to do prediction {end - start}")

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2  = r2_score(y_test, y_pred)

print(f"R² (score): {r2:.4f}")
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
tree = DecisionTreeRegressor(random_state=42)

param_grid_tree = {
    "max_depth": [None, 5, 10, 20],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}

grid_tree = GridSearchCV(tree, param_grid_tree, cv=3, scoring="neg_mean_squared_error", n_jobs=-1)
grid_tree.fit(X_train_scaled, y_train)

print("Best params (DecisionTree, GridSearch):", grid_tree.best_params_)
print("MSE (DecisionTree):", mean_squared_error(y_test, grid_tree.best_estimator_.predict(X_test_scaled)))
```

```python
import time

start = time.time()
y_pred = grid_tree.best_estimator_.predict(X_test)
end = time.time()
inference_time = (end - start) / len(X_test_scaled)

print(f"time taken to do prediction {end - start}")

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2  = r2_score(y_test, y_pred)

print(f"R² (score): {r2:.4f}")
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
```

### Random forest


```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor(random_state=42)
```

```python
param_grid_forest = {
    "n_estimators": [2, 5],
    "max_depth": [None, 10, 20],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}

grid_forest = GridSearchCV(forest, param_grid_forest, cv=3, scoring="neg_mean_squared_error", n_jobs=-1)
grid_forest.fit(X_train, y_train)

print("Best params (RandomForest, GridSearch):", grid_forest.best_params_)
print("MSE (RandomForest, GridSearch):", mean_squared_error(y_test, grid_forest.best_estimator_.predict(X_test)))
```

```python
import time

start = time.time()
y_pred = grid_forest.best_estimator_.predict(X_test)
end = time.time()
inference_time = (end - start) / len(X_test_scaled)

print(f"time taken to do prediction {end - start}")

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2  = r2_score(y_test, y_pred)

print(f"R² (score): {r2:.4f}")
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
```

### Graphical representation

```python
import matplotlib.pyplot as plt
import numpy as np


y_true = y_test.values
y_pred = y_pred  
cols = ["A0", "A1", "A2"]

fig, axes = plt.subplots(2, 3, figsize=(15, 8))

for i, col in enumerate(cols):
    # --- Scatter real vs predicho ---
    axes[0, i].scatter(y_true[:, i], y_pred[:, i], alpha=0.7, color="blue")
    axes[0, i].plot([y_true[:, i].min(), y_true[:, i].max()],
                    [y_true[:, i].min(), y_true[:, i].max()],
                    "r--", lw=2, label="Ideal")
    axes[0, i].set_title(f"{col}: Real vs Predict")
    axes[0, i].set_xlabel("Real")
    axes[0, i].set_ylabel("Predict")
    axes[0, i].legend()
    axes[0, i].grid(True)

    # --- Residuos ---
    residuals = y_true[:, i] - y_pred[:, i]
    axes[1, i].scatter(y_true[:, i], residuals, alpha=0.7, color="green")
    axes[1, i].axhline(0, color="red", linestyle="--")
    axes[1, i].set_title(f"{col}: Fails")
    axes[1, i].set_xlabel("Real")
    axes[1, i].set_ylabel("Error")
    axes[1, i].grid(True)

plt.tight_layout()
plt.show()
```

### Conclusion

Maybe the approach D0 is the best, do A0,A1,A2 approach doesn't add value to the final product. We had already a good algortihm doing the prediction over D0 and assuming k as constant.
